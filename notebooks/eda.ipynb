```python
import pandas as pd

df = pd.read_csv("../data/customer_data.csv")

df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt

df["Age"].plot.box()

plt.show()
```

```python
df[df["Age"] > 100]
```

```python
df = df[df["Age"] <= 100]
```

```python
df["Age"].max()
```

```python
missing_percentage = (df.isnull().sum() / len(df)) * 100

missing_percentage.sort_values(ascending=False)
```

```python
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns

numeric_cols
```

```python
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())
```

```python
df[numeric_cols].isnull().sum()
```

```python
categorical_cols = df.select_dtypes(include=['object']).columns

categorical_cols
```

```python
df[categorical_cols] = df[categorical_cols].fillna(df[categorical_cols].mode().iloc[0])
```

```python
df.isnull().sum()
```

```python
df.to_csv("../data/cleaned_customer_data.csv", index=False)
```

Most customers have lower average order values, while a small number of customers show very high spending behavior, creating a right-skewed distribution.
